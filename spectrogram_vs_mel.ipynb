```python
wav=librosa.load('/home/dan/Downloads/OSR_us_000_0010_8k.wav',sr=8000)
```

```python
wav[0].shape
```

```python
msp = np.abs(librosa.core.stft(wav[0]))
plt.pcolormesh(range(msp.shape[1]), range(msp.shape[0]),10*librosa.amplitude_to_db(msp))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
msp = np.abs(librosa.core.stft(wav[0]))
```

```python
librosa.display.specshow(librosa.amplitude_to_db(msp,
...                                                  ref=np.max),
...                          y_axis='hz', x_axis='time')
```

```python
msp = librosa.feature.mfcc(wav[0],sr=8000,n_mfcc=13)
```

```python
librosa.display.specshow(librosa.amplitude_to_db(msp,
...                                                  ref=np.max),
...                          y_axis='hz', x_axis='time')
```

```python
librosa.display.specshow(librosa.amplitude_to_db(msp,
...                                                  ref=np.max),
...                          y_axis='log', x_axis='time')
```
